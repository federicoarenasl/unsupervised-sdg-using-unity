# Baseline training progress notebook
This notebook is used to analyse the training progress of the baseline under rotation, and rotation and translation experiments. We will analyse the loss of the Distribution Transformer, so the Graph Convolutional Network, the MMD loss, and the downstream task's accuracy progress during the training of the meta-sim network for MNIST data generation.


### Basic library import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

## Results from MNIST rotation

```python
# Get relative paths
rotation_path = '../logs/progress/mnist_rotation'
exp_path = 'exp_1'
gcn_data = os.path.join(rotation_path, exp_path, "gcn.csv")
task_data = os.path.join(rotation_path, exp_path, "task.csv")
# Get dataframes
gcn_data = pd.read_csv(gcn_data, index_col = False)
task_data = pd.read_csv(task_data, index_col = False)
```

### Results for training the Graph Convolutional Network

```python
gcn_data
```

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(12,6))
plt.plot(gcn_data['rec_epoch'], gcn_data['class_loss'], label='class loss')
plt.plot(gcn_data['rec_epoch'], gcn_data['cont_loss'], label='cont loss')
plt.plot(gcn_data['rec_epoch'], gcn_data['loss'], label='(total) loss')
plt.legend(fontsize=12)
plt.title("GCN train loss on Rotation", fontweight='bold', fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Losses", fontsize=12)
plt.show()
```

### Results for training the Meta-Sim model

```python
task_data.tail()
```

### Accuracy

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(10,5))
plt.plot(task_data['epoch'], task_data['task_acc'], label='task accuracy')
plt.title("Meta-Sim downstream task Accuracy on Rotation", fontweight='bold', fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.show()
```

### Losses

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(12,6))
plt.plot(task_data['epoch'], task_data['mmd_loss'], label='MMD loss', alpha=0.7)
plt.plot(task_data['epoch'], task_data['task_loss'], label='task loss', alpha=0.7)
#plt.plot(task_data['epoch'], task_data['loss'], label='(total) loss', alpha=0.7)
plt.legend(fontsize=12)
plt.title("Meta-Sim train losses on Rotation", fontweight='bold', fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.show()
```

# Graph Generation Training Progress
In this part of the notebook we will investigate the effect of adding more nodes and attributes on the total loss of the GCN. We hypothesize that the more features and nodes the GCN needs to predict, the higher the loss will be, and the harder it will be to make it converge to zero in a finite amount of training epochs.

```python
# Get relative paths
rotation_path = '../logs/progress/tyol_rot'
gcn_data = [pd.read_csv(os.path.join(rotation_path, 'exp_'+str(exp_num), "gcn.csv"), index_col = False) for exp_num in range(13, 19)]
```

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(12,10))
labels = ['3 nodes, 3 attrs', '3 nodes, 4 attrs', '5 nodes, 8 attrs', '6 nodes, 12 attrs', '6 nodes, 15 attrs',
         '7 nodes, 20 attrs']
for i,data in enumerate(gcn_data):
    plt.plot(data['rec_epoch'], data['loss'], label=labels[i])
plt.legend(fontsize=15)
plt.title("GCN train loss on increasing Complexity", fontweight='bold', fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Losses", fontsize=12)
plt.ylim(0, 0.15)
plt.xlim(5, 25)
plt.show()
```
